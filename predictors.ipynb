```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps.head()
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
```

```python
del players["Rk"]
```

```python
players["Player"] = players["Player"].str.replace("*","", regex=False)
```

```python
players.head()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players.head()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined[combined["Pts Won"] > 0]
```

```python
combined
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy()
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
sorted(teams["Team"].unique())
```

```python
sorted(combined["Tm"].unique())
```

```python
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
train = combined.merge(teams, how="outer",on=["Team", "Year"])
```

```python
train
```

```python
del train["Unnamed: 0"]
```

```python
train.dtypes
```

```python
train = train.apply(pd.to_numeric, errors='ignore')
```

```python
train.dtypes
```

```python
train["GB"].unique()
```

```python
train["GB"] = pd.to_numeric(train["GB"].str.replace("—","0"))
```

```python
train["GB"].unique()
```

```python
position_mvp_share = train.groupby('Pos')['Share'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
position_mvp_share.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("Set2", len(position_mvp_share)), legend=False)
plt.title("MVP Share by Position")
plt.ylabel('')  # Remove the y-axis label
plt.show()
```

```python
# MVP Share trend over the years
mvp_trend = train.groupby('Year')['Share'].mean()
plt.figure(figsize=(10, 5))
mvp_trend.plot(marker='o')
plt.title("Average MVP Share Over the Years")
plt.xlabel("Year")
plt.ylabel("Average MVP Share")
plt.show()
```

```python
# Bar Chart for Average Points per Team
team_points = train.groupby('Tm')['PTS'].mean().sort_values()
team_points.plot(kind='bar', figsize=(12, 8), color='orange')
plt.title("Average Points per Team")
plt.xlabel("Team")
plt.ylabel("Average Points")
plt.show()
```

```python
team_yearly_points = train.groupby(['Year', 'Tm'])['PTS'].mean()
highest_points_team_per_year = team_yearly_points.groupby('Year').idxmax()
highest_points_team_data = team_yearly_points[highest_points_team_per_year]
print(highest_points_team_data)

highest_points_team_data.plot(kind='bar', figsize=(12, 8), color='green')
plt.title("Team with Highest Average Points per Year")
plt.xlabel("Year")
plt.ylabel("Average Points")
plt.show()
```

```python
# Find the player with the most 3-pointers
player_3p = train.groupby('Player')['3P'].sum()
most_3p_player = player_3p.idxmax()
most_3p_value = player_3p.max()

# Find the player with the most 2-pointers
player_2p = train.groupby('Player')['2P'].sum()
most_2p_player = player_2p.idxmax()
most_2p_value = player_2p.max()

# Find the player with the most field goals (FG)
player_fg = train.groupby('Player')['FG'].sum()
most_fg_player = player_fg.idxmax()
most_fg_value = player_fg.max()

# Find the player with the most free throws (FT)
player_ft = train.groupby('Player')['FT'].sum()
most_ft_player = player_ft.idxmax()
most_ft_value = player_ft.max()

# Find the player with the most total points (PTS)
player_pts = train.groupby('Player')['PTS'].sum()
most_pts_player = player_pts.idxmax()
most_pts_value = player_pts.max()

# Find the player with the most assists (AST)
player_ast = train.groupby('Player')['AST'].sum()
most_ast_player = player_ast.idxmax()
most_ast_value = player_ast.max()

# Print the results
print(f"Player with most 3-pointers: {most_3p_player} ({most_3p_value} 3P)")
print(f"Player with most 2-pointers: {most_2p_player} ({most_2p_value} 2P)")
print(f"Player with most field goals: {most_fg_player} ({most_fg_value} FG)")
print(f"Player with most free throws: {most_ft_player} ({most_ft_value} FT)")
print(f"Player with most total points: {most_pts_player} ({most_pts_value} PTS)")
print(f"Player with most assists: {most_ast_player} ({most_ast_value} AST)")
```

```python
team_win_percentage = train.groupby('Tm')['W/L%'].mean()
highest_win_percentage_team = team_win_percentage.idxmax()
highest_win_percentage_value = team_win_percentage.max()

print(f"Team with highest winning percentage: {highest_win_percentage_team}")
print(f"Winning percentage: {highest_win_percentage_value:.2f}")
```

```python
min_age_player = train.loc[train['Age'].idxmin()]['Player']
min_age_value = train['Age'].min()

max_age_player = train.loc[train['Age'].idxmax()]['Player']
max_age_value = train['Age'].max()

print(f"Youngest player: {min_age_player} (Age: {min_age_value})")
print(f"Oldest player: {max_age_player} (Age: {max_age_value})")
```

```python
train['PER'] = (train['PTS'] + (train['AST'] * 0.5) + (train['TRB'] * 0.7) +
                (train['STL'] * 0.8) + (train['BLK'] * 0.8) -
                (train['TOV'] * 1) - (train['PF'] * 0.7)) / train['G']

best_per_player = train.groupby('Player')['PER'].mean().idxmax()
best_per_value = train.groupby('Player')['PER'].mean().max()

print(f"Player with Best PER: {best_per_player} ({best_per_value:.2f})")
```

```python
team_offensive_defensive = train.groupby('Tm')[['PS/G', 'PA/G']].mean()
best_offense_team = team_offensive_defensive['PS/G'].idxmax()
best_defense_team = team_offensive_defensive['PA/G'].idxmin()

print(f"Best Offensive Team: {best_offense_team}")
print(f"Best Defensive Team: {best_defense_team}")
```

```python
age_performance_corr = train[['Age', 'PTS', 'AST', 'TRB', 'STL']].corr()
print(age_performance_corr)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(age_performance_corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Age and Performance Metrics')
plt.show()
```

```python
# Group by Player and Year to calculate total points per player per year
player_season_pts = train.groupby(['Player', 'Year'])['PTS'].sum().unstack()

# Iterate over each player and generate a report and plot
for player in player_season_pts.index:
    player_data = player_season_pts.loc[player]
    
    # Plot the player's points over the years
    plt.figure(figsize=(10, 6))
    player_data.plot(kind='line', marker='o', linestyle='-', title=f"{player} Points Per Year")
    plt.ylabel("Total Points")
    plt.xlabel("Year")
    
    # Display the chart
    plt.grid(True)
    plt.show()
```
